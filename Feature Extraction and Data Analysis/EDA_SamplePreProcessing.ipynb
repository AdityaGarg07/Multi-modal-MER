```python
# Function to normalize a Eda Song data
def normalizedf(df):
    normalized_df=(df-df.mean())/df.std()
    normalized_df['time(s)'] = df['time(s)']
    return normalized_df
```

```python
# Normalization of all the available songs in the Dataset
path = '/Users/aditya_mac/Downloads/PMEmo2019/EDA/*.csv'
pathx = '/Users/aditya_mac/Downloads/PMEmo2019/NewEDA'
for fname in sorted(glob.glob(path)):
    newfile = (pathx + fname[41:-4] + 'n' + '.csv')
    df = pd.read_csv(fname)
    newdf = normalizedf(df)
    newdf.iloc[:,0:11].to_csv(newfile,index = False)
```
